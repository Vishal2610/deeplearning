```python
import tensorflow as tf
import tensorflow.contrib.layers as layers
from tensorflow.examples.tutorials.mnist import input_data
```

```python
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
```

```python
batchSz = 100
```

```python
img = tf.placeholder(tf.float32, [batchSz, 784])
ans = tf.placeholder(tf.float32, [batchSz, 10])
```

```python
# U = tf.Variable(tf.random_normal([784, 784], stddev=.1))
# bU = tf.Variable(tf.random_normal([784], stddev=.1))
# V = tf.Variable(tf.random_normal([784, 10], stddev=.1))
# bV = tf.Variable(tf.random_normal([10], stddev=.1))

# l1_output = tf.matmul(img, U) + bU
# l1_output = tf.nn.relu(l1_output)
# prbs = tf.nn.softmax(tf.matmul((l1_output, V) + bV))
```

```python
l1_output = layers.fully_connected(img, 756)
prbs = layers.fully_connected(l1_output, 10, tf.nn.softmax)
```

```python
xEnt = tf.reduce_mean(-tf.reduce_sum(ans*tf.log(prbs), reduction_indices=[1]))
```

```python
train = tf.train.GradientDescentOptimizer(0.05).minimize(xEnt)
```

```python
num_correct = tf.equal(tf.argmax(prbs, 1), tf.argmax(ans, 1))
```

```python
accuracy = tf.reduce_mean(tf.cast(num_correct, tf.float32))
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for i in range(4000):
    imgs, anss = mnist.train.next_batch(batchSz)
    if(i%10 == 0):
        acc, ignore = sess.run([accuracy, train], feed_dict={img: imgs, ans: anss})
        print('Iteration: ', i, 'Accuracy: ', acc*100)
```

```python
sumAcc = 0
for i in range(4000):
    imgs, anss = mnist.test.next_batch(batchSz)
    sumAcc += sess.run(accuracy, feed_dict={img: imgs, ans: anss})
print('Accuracy: ', sumAcc/4000*100)
```
